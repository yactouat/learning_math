# Plotting figures

Plotting figures on the coordinate plane is an important skill to have if you want to be able to perform transformations using matrices. In this section, we will learn how to plot figures on the coordinate plane using Python.

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Plotting a square

To plot a figure, you need to know the coordinates of its vertices. For example, to plot a square, you need to know the coordinates of its four vertices. Let's say that the square has vertices at $(1,1)$, $(1,3)$, $(3,3)$, and $(3,1)$. To plot this square, we can use the following code:

```python
# initializing the figure
plt.figure(figsize=(6, 6))

# defining the x vertices of our square,
# we repeat the first vertex at the end to close the square
x = [1, 1, 3, 3, 1] 

# defining the y vertices of our square
y = [1, 3, 3, 1, 1]

# plot the square,
# in blue color with 'o' markers at vertices
plt.plot(x, y, 'b-o')

# add labels to the vertices
for i in range(len(x)):
    plt.text(x[i], y[i], f'({x[i]}, {y[i]})')

# add a grid to the plot
plt.grid(True)

# set the x and y axis limits
plt.xlim(0, 6)
plt.ylim(0, 6)

# add a title and axis labels
plt.title('Plotting a square')
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```
