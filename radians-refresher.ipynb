# Radians refresher

Here is a visual representation that helps you remember what a radian is.

Radians are defined in terms of the radius of a circle.

```python
# let's import the necessary libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the circle and its properties
circle_radius = 1
theta = np.linspace(0, 2*np.pi, 300)
x_circle = circle_radius * np.cos(theta)
y_circle = circle_radius * np.sin(theta)

# Arc for 1 radian
theta_radian = np.linspace(0, 1, 100)
x_radian = circle_radius * np.cos(theta_radian)
y_radian = circle_radius * np.sin(theta_radian)

# Plotting
plt.figure(figsize=(8,8))
plt.plot(x_circle, y_circle, label='Circle', color='blue')
plt.plot(x_radian, y_radian, label='1 Radian Arc', color='green', linewidth=2)

# Plot the radius
plt.plot([0, 1], [0, 0], color='black', linestyle='--', label='Radius')

# Annotations
plt.annotate('1 Radian', (0.5, 0.1), fontsize=9, ha='center')

# Setting aspect ratio, limits, and labels
plt.gca().set_aspect('equal', adjustable='box')
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Radians Visualization')
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()

plt.show()
```
