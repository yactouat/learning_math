# `sin` and `cos` functions

The _sine_ and _cosine_ functions come from _trigonometry_. These functions are used in many areas of mathematics, science, and engineering. For instance, they are used a lot in everything that includes _triangulation_ (e.g. determining unknown distances or positions): GPS technology, astronomy, surveying, etc.

Trigonometry is a branch of mathematics that deals with relationships between the sides and angles of triangles.

The sine and consine functions are defined using a _unit circle_.

A unit circle is a circle with a _radius_ of 1.

The radius of a circle is the distance from the center of the circle to any point on the circle's circumference. Imagine a circle as a pizza, and the radius is the length of a straight slice from the center of the pizza to its edge.

```python
# let's import the necessary libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# let's illustrate the notions of radius and unit circle

# Create a circle with radius 2 and a unit circle with radius 1
circle_radius_2 = plt.Circle((0, 0), 2, color='blue', fill=False, label='Circle (radius=2)')
unit_circle = plt.Circle((0, 0), 1, color='green', fill=False, label='Unit Circle (radius=1)')

# Plot the circles
fig, ax = plt.subplots(figsize=(7,7))
ax.add_artist(circle_radius_2)
ax.add_artist(unit_circle)

# Set the limits and labels
plt.xlim(-2.5, 2.5)
plt.ylim(-2.5, 2.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Circles with Radii')
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)

# Draw the radii
plt.plot([0, 2], [0, 0], color='blue', linestyle='--', label='Radius (length=2)')
plt.plot([0, np.cos(np.pi/4)], [0, np.sin(np.pi/4)], color='green', linestyle='--', label='Radius (length=1)')

# Add a legend
plt.legend()

plt.show()
```

As you can see in the example above, the unit circle is the one that has a radius of 1. This will be important as we'll dive into what sine and cosine functions are.

## Definition of Sine and Cosine

Sine and cosine are defined as follows:

- The cosine of an angle is the the x-coordinate of the point where the angle intersects the unit circle.
- The sine of the same angle is the y-coordinate of the point where the angle intersects the unit circle.

Using a unit circle with has it's center at the origin of a Cartesian coordinate plan, we can illustrate this as follows =>

```python
# Create a circle
circle = plt.Circle((0, 0), 1, color='b', fill=False)

# Plot the circle
fig, ax = plt.subplots(figsize=(5,5))
ax.add_artist(circle)

# Set the limits and labels
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.xlabel('X')
plt.ylabel('Y')

# Draw the x and y axes
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)

# Draw the angle
angle = np.deg2rad(30)  # Angle in radians
x = np.cos(angle)
y = np.sin(angle)

# Draw the line from the origin to the point (x,y)
plt.plot([0, x], [0, y], color='r')

# Mark the point on the circumference
plt.plot(x, y, 'ro')

# Add annotations for sine and cosine
plt.annotate(f'cos(30°) = {x:.2f}', (x/2, -0.2), textcoords="offset points", xytext=(10,-10), ha='center', fontsize=9)
plt.annotate(f'sin(30°) = {y:.2f}', (x + 0.2, y/2), textcoords="offset points", xytext=(10,10), ha='center', fontsize=9)

plt.show()
```

If we were to plot sin and cosine over a range of angles (from 0 to 360 degrees), we would see a familiar pattern, that repeats every 360 degrees... This is why sine and cosine are called periodic functions.

```python
# Create an array of angles from 0 to 2*pi (which is equivalent to 0 to 360 degrees)
angles = np.linspace(0, 2*np.pi, 100)

# Compute the corresponding sine and cosine values
sine_values = np.sin(angles)
cosine_values = np.cos(angles)

# Create the plot
plt.figure(figsize=(10, 6))

# Plot sine values
plt.plot(angles, sine_values, label='sin')

# Plot cosine values
plt.plot(angles, cosine_values, label='cos')

# Set the limits and labels
plt.xlim(0, 2*np.pi)
plt.ylim(-1.5, 1.5)
plt.xlabel('Angle (radians)')
plt.ylabel('Value')
plt.title('Sine and Cosine Functions')
plt.grid(True)

# Add a legend
plt.legend()

plt.show()
```
