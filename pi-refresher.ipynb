# Pi refresher

π (pi) is a mathematical constant that represents the ratio of a circle's circumference (the distance around the circle) to its diameter (the distance across the circle at its widest point).

Whatever the circle, this ratio is always the same:  approximately 3.14159...

π is an irrational number, which means that it cannot be expressed exactly as a ratio of two integers (such as 22/7 or other fractions that are commonly used to approximate π); consequently, its decimal representation never ends and never settles into a permanent repeating pattern.

π is used in various mathematical, scientific, and engineering contexts, especially when dealing with circles or circular motion.

```python
# let's import the necessary libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the circle and its properties
circle_radius = 1
theta = np.linspace(0, 2*np.pi, 300)
x_circle = circle_radius * np.cos(theta)
y_circle = circle_radius * np.sin(theta)

# Plotting
plt.figure(figsize=(6,6))
plt.plot(x_circle, y_circle, label='Circumference', color='blue')

# Plot the diameter
plt.plot([-1, 1], [0, 0], color='red', linestyle='--', label='Diameter')

# Annotations
plt.annotate('Diameter', (0, -0.1), fontsize=9, ha='center', va='top')
plt.annotate('Circumference is approximately\n3.14159 times the diameter', (0, 0.75), fontsize=9, ha='center')

# Setting aspect ratio, limits, and labels
plt.gca().set_aspect('equal', adjustable='box')
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Visualization of π (pi)')
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()

plt.show()
```

Pi can be used to calculate the circumference of a circle, as in:

$$C = 2\pi r$$

where $r$ is the radius of the circle.
