```python
from fastai.basics import *

# configuring a function to plot the quadratic function
plt.rc('figure', dpi=90)
def plot_function(f, title=None, min=-2.1, max=2.1, color='r', ylim=None):
    x = torch.linspace(min,max, 100)[:,None]
    if ylim: plt.ylim(ylim)
    plt.plot(x, f(x), color)
    if title is not None: plt.title(title)
```

A quadratic function is a type of mathematical function that has the form:

`f(x) = ax² + bx + c`

... where `a`, `b`, and `c` are constants. These constants determine the _shape_ and _position_ of the function. They can be any real number.

The graph of a quadratic function is a parabola (a _U_ shape or an upside down _U_ shape). 

- If _a_ is positive, the parabola opens upwards, else it opens downwards.
- _b_ will shift the parabola left or right.
- _c_ will shift the parabola up or down.

The _vertex_ of the parabola is the point where the parabola is at its minimum or maximum value. To find the x-coordinate of the vertex of a quadratic function, use the formula: 

`xc = -b / 2a`

... and to find the y-coordinate of the vertex, substitute the x-coordinate into the function. As in:

`y = f(xc) = axc² + bxc + c`

```python
# let's define a quadratic function in Python
def quadratic(x, a, b, c): return a*(x**2) + b*x + c

def print_vertex_coordinates(a, b, c):
    # let's print the vertex of the quadratic function
    print(f'vertex x = {-b/(2*a)}')
    print(f'vertex y = {quadratic((-b/(2*a)), a, b, c)}')
```

now let's see some examples of quadratic functions with different values of _a_, _b_, and _c_ etc.

The _axis of symmetry_ is the line that divides the parabola into two equal halves. It is the line that passes through the vertex of the parabola. The x-coordinate of the vertex is the equation of the axis of symmetry.

Now, let's talk about solutions of a quadratic function; these solutions are where the distinct x-values where the graph of the function crosses the x-axis. These are also called the _roots_ of the function.

```python
# this is how we would solve the quadratic function in Python
def find_quadratic_solutions(a, b, c):
    # let's find the solutions of the quadratic function
    solution1 = (-b + math.sqrt((b**2) - (4*a*c))) / (2*a)
    solution2 = (-b - math.sqrt((b**2) - (4*a*c))) / (2*a)
    print(f'solutions = {solution1}, {solution2}')

# the (b**2) - (4*a*c) part is called the discriminant
```

```python
def ex8(x): return quadratic(x, 0, 0, 0)
plot_function(ex8, "$ax^2 + bx + c$")
# here, there is one solution for every x, which is 0
```

```python
# let's reuse a previous example and find the solutions of the quadratic function
def ex52(x): return quadratic(x, 1, -1, 0)
plot_function(ex52, "$ax^2 + bx + c$")
```

```python
find_quadratic_solutions(1, -1, 0)
try:
    find_quadratic_solutions(1, 0, 0)
except:
    print('There is no solution for this quadratic function')
```

The _Y-Intercept_ of a quadratic function is where the graph of the function crosses the y-axis. The y-intercept is the value of the function when x is zero. So, to find the y-intercept, substitute zero for x in the function. Eventually, you will get the value of _c_ at point `(0, c)`.

## Practical real-world example

Revenue for a business is expressed as the product of a single product unit price and the number of units sold.

For a given product, if the price is too high, you won't have as many customers; if the price is too low, there won't be much profit per sold item. This can be expressed, with `N` being the number of units you can sell, `S` being the number of items you have in stock, and `P` being the price of the product, as: `N = S - P`. This is a linear function that suggests that, as the price increases, the number of units sold decreases linearly.

What you want is to find the optimal price for the product in order to maximize revenue. Revenue `R` can be expressed as can be expressed as _the price per unit times the number of units sold_: `R = P * N`. We need to find the price that gives the highest value of `R`, in other words the vertex of an inversed _U_ described by the quadratic function.

First, let's turn `R = P * N` into => `R = P * (S - P)` => `R = SP - P²` => `R = -P² + SP` => this is a quadratic function in the variable `P` (the price). Since `P` is a negative value here, the parabola opens downwards. The vertex will represent the maximum value of the function.

```python
# let's plot this, -1 being the coefficient of `P`
def practicalEx(x): return quadratic(x, -1, 7.5, 0)
def plot_practical_function(f, title=None, min=-15.1, max=100.1, color='r', ylim=None):
    x = torch.linspace(min,max, 100)[:,None]
    if ylim: plt.ylim(ylim)
    plt.plot(x, f(x), color)
    if title is not None: plt.title(title)
plot_practical_function(practicalEx, "$ax^2 + bx + c$")
find_quadratic_solutions(-1, 7.5, 0)
```
