# Calculus: the notion of a slope

The slope a line represents the rate at which `y` changes for each unit change in `x`.

The slope of a line is a measure of how steep the line is. It is calculated by finding the change in the vertical values (usually denoted as `y`) divided by change in the horizontal values (usually denoted as `x`). This is often referred to as _rise over run_.

If we have two points on a line `(x1, y1)` and `(x2, y2)`, the slope of the line is `(y2 - y1) / (x2 - x1)`.

```python
# let's plot a simple line that has the equation `y = x + 1`
import matplotlib.pyplot as plt
import numpy as np

# Define the line
def line(x):
    return x + 1

# create an array of x values
# the `np.linspace` function creates an array of 50 (3rd default parameter of the `linspace` function) evenly spaced values,
# these values are in the space between 0 and 4, 
# and are organised into a numpy array that has a shape of `(50,)`
x = np.linspace(0, 4)

# Create the plot
plt.figure(figsize=(8, 6))

# the `line` function takes an input `x` array and returns an output array of the same shape, 
# where each element of the output array is the corresponding element of the input array plus 1 
plt.plot(x, line(x), label="y = x + 1")

plt.legend()
plt.grid(True)
plt.title("Plotting `y = x + 1`")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
# now let's do the same, with the goal of calculating the slope of a line that passes through the points (1, 2) and (3, 4)

# Define the points
x_points = [1, 3]
y_points = [2, 4] 

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, line(x), label="y = x + 1")
plt.plot(x_points, y_points, 'ro')  # plot the points
plt.text(x_points[0], y_points[0], '  (1,2)', verticalalignment='bottom', fontsize=12, color='red')
plt.text(x_points[1], y_points[1], '  (3,4)', verticalalignment='bottom', fontsize=12, color='red')
plt.legend()
plt.grid(True)
plt.title("Plot of a straight line with calculated slope")
plt.xlabel("x")
plt.ylabel("y")
plt.show()

def calculate_slope(x1, y1, x2, y2):
    return (y2 - y1) / (x2 - x1)

slope = calculate_slope(x_points[0], y_points[0], x_points[1], y_points[1])
print("The slope of the line is {}".format(slope))
```

## the equation of a straight line

The equation of a straight line can be expressed as `y = mx + b` where `m` is the slope and `b` is the y-intercept. 

The slope is the change in y divided by the change in x. 

The y-intercept is the value of y when x is zero.

`y` would be the _dependent variable_, the one that changes based on the value of `x`, the _independent variable_. In deep learning, we often use `y` to represent the label, or the value we are trying to predict.

```python
# Define a function for the line
def line(x, m, b):
    return (m * x) + b

# Set values for the slope and y-intercept
    # `m` represents the rate at which `y` changes for each unit change in `x` 
m = 2
b = 3

# Generate x values
x = np.linspace(-1, 5, 400)

# Generate y values for the line
y = line(x, m, b)

# Create the plot
plt.figure(figsize=(8, 6))

# Plot the line
plt.plot(x, y, label='y = {}x + {}'.format(m, b))

# Highlight the y-intercept, `ro` means red circle marker
plt.plot(0, b, 'ro')
plt.text(0, b, '  y-intercept ({}, {})'.format(0, b))

# Draw a triangle to illustrate the slope by plotting a green line
    # `[1, 1, 2]` is the x coordinates of the triangle corners
    # `[m*1 + b, m*2 + b, m*2 + b]` is the y coordinates of the triangle corners
    # `'g-'` means a solid green line
# this triangle represents the "rise over run" (or slope) concept
plt.plot([1, 1, 2], [m*1 + b, m*2 + b, m*2 + b], 'g-')
plt.text(1.5, m*1.5 + b, 'slope = {}'.format(m))

# plotting `y` as `x` increases by 1
plt.text(1,5, ' (1,5)', verticalalignment='bottom', fontsize=12, color='red')
plt.text(2,7, ' (2,7)', verticalalignment='bottom', fontsize=12, color='red')

# Add labels and a legend
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

## point-slope form of a linear equation

The slope `m` of a line passing through points `(x1, y1)` and `(x, y)` is given by:

`m = (y - y1) / (x - x1)`

This can be re written as:

`m * (x - x1) = y - y1`
